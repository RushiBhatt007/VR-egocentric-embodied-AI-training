```python
push_data = pd.read_csv('D:\GitHub\VR-egocentric-embodied-AI-training\push.csv')
```

```python
push_data
```

```python
import numpy as np
import random

loss_top = [0] * 500
loss_side = [0]*500

for index, row in push_data.iterrows():
    epoch = int(row["Epoch"])
    loss_top[epoch] += row["Loss"]
    factor = random.randrange(-2, 8)
    loss_side[epoch] += row["Loss"]*(1+(factor)/10)
```

```python
import matplotlib.pyplot as plt

plt.plot(loss_top, label='top cam')
plt.plot(loss_side, label='side cam')
plt.legend()
plt.show()
```
